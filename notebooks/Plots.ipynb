```python
%matplotlib inline

import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%load_ext autoreload
%autoreload 2
```

## Beacon Agent

```python
beacon_folder = r'../sc2bot/agents/data/MoveToBeacon/gcp_trained'
fn_header = r'/10000eps_BeaconAgent_data_'
pretrained_tail = 'pretrain.pkl'
untrained_tail = 'no_pretrain.pkl'
pretrained_fn = beacon_folder + fn_header + pretrained_tail
untrained_fn = beacon_folder + fn_header + untrained_tail
pretrained = pickle.load(open(pretrained_fn, 'rb'))
untrained = pickle.load(open(untrained_fn, 'rb'))
pretrained.keys()
```

### Loss

```python
plt.figure(figsize=(16, 10))
plt.plot(pretrained['loss'])
plt.plot(untrained['loss'])
plt.xlabel('Number of Episodes')
plt.ylabel('Loss (MSE)')
plt.legend(['pretrained', 'not pretrained'])
plt.title('Loss for DQNs on Beacon Task with and without pretraining from experience')
None
```

### Reward

```python
plt.figure(figsize=(16, 12))
plt.plot(pretrained['reward'])
plt.plot(untrained['reward'])
plt.xlabel('Number of Episodes')
plt.ylabel('Reward')
plt.legend(['pretrained', 'not pretrained'])
plt.title('Reward for DQNs on Beacon Task with and without pretraining from experience')
```

### max_q

```python
plt.figure(figsize=(16, 10))
plt.plot(pretrained['max_q'])
plt.plot(untrained['max_q'])
plt.xlabel('Number of Episodes')
plt.ylabel('Max q-value')
plt.legend(['pretrained', 'not pretrained'])
plt.title('Max q-values for DQNs on Beacon Task with and without pretraining from experience')
None
```
